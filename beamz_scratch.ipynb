```python
import beamz9000 as beamz
from dataclasses import dataclass
from typing import Union
from matplotlib.patches import PathPatch
```

```python
n01 = beamz.Node(0, "A")
n02 = beamz.Node(5.5, "B")
n03 = beamz.Node(15, "D")
n04 = beamz.Node(14.5, "C")

pin = beamz.Support(n01, 2)
roller = beamz.Support(n02, 0)
fix = beamz.Support(n04, 3)

udl = beamz.Load(200, n01, n02)

beam = beamz.Beam(
    nodes=[n01, n02, n04, n03], 
    supports=[pin, roller, fix], 
    loads=udl,
    dimensions=[n01, n02], 
    depth=0.25)
```

```python
import matplotlib.transforms as transforms

scale = transforms.Affine2D()
scalet = scale.scale(4)
scalet.transform([1., 1.,])
```

```python
bp = beamz.BeamPlotter(beam)
bp.plot()
```
